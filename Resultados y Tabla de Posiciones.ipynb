```python
import pandas as pd
import numpy as np
import math
```

```python
data_location = 'output/1ra. B AFA 1960.csv'

data = pd.read_csv(data_location)
#pd.options.display.max_rows = None
data.head(15)
```

```python
HTML_results=''
data['Goles1']=data['Goles1'].astype(str)
data['Goles2']=data['Goles2'].astype(str)
nfant=0
for np, i in enumerate(data.index):
    cant=len(data['Equipo1'].unique())
    cpf=math.floor(cant/2)
    nf=(np+1)/cpf
    nf=math.ceil(nf)
    nf=int(nf)
    if nf!=nfant:
        if nf>1:         
            HTML_results+='|}\n'
        HTML_results+='{|class="wikitable collapsible collapsed" style="font-size:95%; text-align: center; width: 95%;"\n!colspan="12" style="with: 100%;" align=center | Fecha '+f'{nf}\n'+'|-\n!width="150"|Local\n!width="75"|Resultado\n!width="150"|Visitante\n!width="150"|Estadio\n!width="95"|Fecha\n'
        nfant=nf
    HTML_results+='|-align=center\n'+'|'+f"{data.loc[i,'Equipo1']}\n"+'|'+f"{data.loc[i,'Goles1']}"+' - '+f"{data.loc[i,'Goles2']}\n"+'|'+f"{data.loc[i,'Equipo2']}\n"+'|'+f"{data.loc[i,'Lugar']}\n"+'|'+f"{data.loc[i,'Fecha']}\n"
HTML_results+='|}'
print(HTML_results)
```

```python
equipos = pd.unique(data['Equipo1'])
PTS = 0
PJ = 0
PG= 0
PE= 0
PP= 0
GF= 0
GC= 0
df=pd.DataFrame({"Equipo":equipos,
                "Puntos":PTS,
                "PJ":PJ,
                "PG":PG,
                "PE":PE,
                "PP":PP,
                "GF":GF,
                "GC":GC})
df.set_index('Equipo', inplace = True)
for i in data.index:
    if data["Goles1"][i]>data["Goles2"][i]:
        df.loc[data["Equipo1"][i], 'Puntos' ] += 2
        df.loc[data["Equipo1"][i], 'PG' ] += 1
        df.loc[data["Equipo2"][i], 'PP' ] += 1
    if data["Goles1"][i]<data["Goles2"][i]:
        df.loc[data["Equipo2"][i], 'Puntos' ] += 2
        df.loc[data["Equipo2"][i], 'PG' ] += 1
        df.loc[data["Equipo1"][i], 'PP' ] += 1
    if data["Goles1"][i]==data["Goles2"][i]:
        df.loc[data["Equipo1"][i], 'Puntos' ] += 1
        df.loc[data["Equipo2"][i], 'Puntos' ] += 1
        df.loc[data["Equipo1"][i], 'PE' ] += 1
        df.loc[data["Equipo2"][i], 'PE' ] += 1
    df.loc[data["Equipo1"][i], 'PJ' ] += 1
    df.loc[data["Equipo2"][i], 'PJ' ] += 1
    df.loc[data["Equipo1"][i], 'GF' ] += data["Goles1"][i]
    df.loc[data["Equipo2"][i], 'GF' ] += data["Goles2"][i]
    df.loc[data["Equipo1"][i], 'GC' ] += data["Goles2"][i]
    df.loc[data["Equipo2"][i], 'GC' ] += data["Goles1"][i]    
df['Dif']=df['GF']-df['GC']
df=df.sort_values(['Puntos', 'Dif', 'GF'], ascending=False)
df
```
