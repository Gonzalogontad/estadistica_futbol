```python
import pandas as pd
import numpy as np
import math
```

```python
data_location = 'output/1ra. B AFA 1960.csv'

data = pd.read_csv(data_location)
#pd.options.display.max_rows = None
data.head()
```

```python
data=data.replace('Sp. ','', regex=True)
data=data.replace('Central Córdoba','Central Córdoba (R)', regex=True)
data=data.replace('Dep.','Deportivo', regex=True)
data=data.replace('Sarmiento','Sarmiento (J)', regex=True)
data=data.replace('Talleres','Talleres (RE)', regex=True)
e1=data['Equipo1'].unique()
e1.sort()
print(e1)
```

```python
data['Lugar']=data['Lugar'].replace('Floresta', 'All Boys')
data['Lugar']=data['Lugar'].replace('José Ingenieros', '[[Estadio Tres de Febrero|Tres de Febrero]]')
data['Lugar']=data['Lugar'].replace('Banfield', '[[Estadio Florencio Sola|Florencio Sola]]')
data['Lugar']=data['Lugar'].replace('Rosario', '[[Estadio Gabino Sosa|Gabino Sosa]]')
#data['Lugar']=data['Lugar'].replace('Núñez', '[[Estadio Juan Pasquale|Juan Pasquale]]')
data['Lugar']=data['Lugar'].replace('Morón', '[[Estadio Francisco Urbano|Francisco Urbano]]')
data['Lugar']=data['Lugar'].replace('Dock Sud', '[[Club_Sportivo_Dock_Sud#Estadio|De Los Inmigrantes]]')
data['Lugar']=data['Lugar'].replace('Gerli', 'El Porvenir')
data['Lugar']=data['Lugar'].replace('Belgrano', '[[Estadio de Excursionistas|Excursionistas]]')
data['Lugar']=data['Lugar'].replace('Lomas de Zamora', '[[Estadio Eduardo Gallardón|Eduardo Gallardón]]') #Así se llama desde el 8 de marzo de 1980
data['Lugar']=data['Lugar'].replace('Lomas de Zamora', 'Los Andes') #Desde el 28 de septiembre de 1940
data['Lugar']=data['Lugar'].replace('Mataderos', '[[Estadio Nueva Chicago|Nueva Chicago]]')
data['Lugar']=data['Lugar'].replace('Quilmes', '[[Quilmes_Atlético_Club#Estadio_de_Guido_y_Sarmiento|Guido y Sarmiento]]')
data['Lugar']=data['Lugar'].replace('Junín', '[[Estadio Eva Perón|Eva Perón]]')
data['Lugar']=data['Lugar'].replace('Remedios de Escalada', '[[Estadio Pablo Comelli|Timote y Manuel Castro]]')
data['Lugar']=data['Lugar'].replace('Temperley', '[[Estadio Alfredo Beranger|Alfredo Beranger]]')
data['Lugar']=data['Lugar'].replace('Victoria', '[[Estadio José Dellagiovanna|José Dellagiovanna]]')
data['Lugar']=data['Lugar'].replace('Santa Fe', '[[Estadio 15 de Abril|15 de Abril]]')
data['Lugar']=data['Lugar'].replace('Paternal', '[[Estadio Diego Armando Maradona (Argentina)|Argentinos Juniors]]')
data['Lugar']=data['Lugar'].replace('Caballito', '[[Estadio Arquitecto Ricardo Etcheverri|Ferro Carril Oeste]]')
```

```python
data=data.replace('/1950', '', regex=True)
data=data.replace('/1960', '', regex=True)
data=data.replace('/03', ' de marzo', regex=True)
data=data.replace('/04', ' de abril', regex=True)
data=data.replace('/05', ' de mayo', regex=True)
data=data.replace('/06', ' de junio', regex=True)
data=data.replace('/07', ' de julio', regex=True)
data=data.replace('/08', ' de agosto', regex=True)
data=data.replace('/09', ' de septiembre', regex=True)
data=data.replace('/10', ' de octubre', regex=True)
data=data.replace('/11', ' de noviembre', regex=True)
data=data.replace('/12', ' de diciembre', regex=True)
data['Fecha']=data['Fecha'].replace('01', '1', regex=True)
data['Fecha']=data['Fecha'].replace('02', '2', regex=True)
data['Fecha']=data['Fecha'].replace('03', '3', regex=True)
data['Fecha']=data['Fecha'].replace('04', '4', regex=True)
data['Fecha']=data['Fecha'].replace('05', '5', regex=True)
data['Fecha']=data['Fecha'].replace('06', '6', regex=True)
data['Fecha']=data['Fecha'].replace('07', '7', regex=True)
data['Fecha']=data['Fecha'].replace('08', '8', regex=True)
data['Fecha']=data['Fecha'].replace('09', '9', regex=True)
data['Fecha'].unique()
```

```python
HTML_results=''
data['Goles1']=data['Goles1'].astype(str)
data['Goles2']=data['Goles2'].astype(str)
nfant=0
for np, i in enumerate(data.index):
    cant=len(data['Equipo1'].unique())
    cpf=math.floor(cant/2)
    nf=(np+1)/cpf
    nf=math.ceil(nf)
    nf=int(nf)
    if nf!=nfant:
        if nf>1 and nf%2!=0:         
            HTML_results+='|}\n|}\n'
        if nf>1 and nf%2==0:
            HTML_results+='|}\n'
        nfant=nf    
        if nf%2!=0:
            HTML_results+='{| width=100%\n| valign=top width=50% align=left |\n{|class="wikitable collapsible collapsed" style="font-size:95%; text-align: center; width: 95%;"\n!colspan="12" style="with: 100%;" align=center | Fecha '+f'{nf}\n'+'|-\n!width="150"|Local\n!width="75"|Resultado\n!width="150"|Visitante\n!width="150"|Estadio\n!width="95"|Fecha\n'
        if nf%2==0:    
            HTML_results+='| valign=top width=50% align=left |\n{|class="wikitable collapsible collapsed" style="font-size:95%; text-align: center; width: 95%;"\n!colspan="12" style="with: 100%;" align=center | Fecha '+f'{nf}\n'+'|-\n!width="150"|Local\n!width="75"|Resultado\n!width="150"|Visitante\n!width="150"|Estadio\n!width="95"|Fecha\n'
    if data.loc[i,'Goles1']>data.loc[i,'Goles2']:
        HTML_results+='|-align=center\n'+"|bgcolor=#D0E7FF|\'\'\'"+f"{data.loc[i,'Equipo1']}\n"+'|'+f"{data.loc[i,'Goles1']}"+' - '+f"{data.loc[i,'Goles2']}\n"+'|'+f"{data.loc[i,'Equipo2']}\n"+'|'+f"{data.loc[i,'Lugar']}\n"+'|'+f"{data.loc[i,'Fecha']}\n"
    if data.loc[i,'Goles1']<data.loc[i,'Goles2']:
        HTML_results+='|-align=center\n'+'|'+f"{data.loc[i,'Equipo1']}\n"+"|"+f"{data.loc[i,'Goles1']}"+' - '+f"{data.loc[i,'Goles2']}\n"+"|bgcolor=#D0E7FF|\'\'\'"+f"{data.loc[i,'Equipo2']}\n"+'|'+f"{data.loc[i,'Lugar']}\n"+'|'+f"{data.loc[i,'Fecha']}\n"
    if data.loc[i,'Goles1']==data.loc[i,'Goles2']:
        HTML_results+='|-align=center\n'+'|'+f"{data.loc[i,'Equipo1']}\n"+"|bgcolor=#D0E7FF|\'\'\'"+f"{data.loc[i,'Goles1']}"+' - '+f"{data.loc[i,'Goles2']}\n"+"|"+f"{data.loc[i,'Equipo2']}\n"+'|'+f"{data.loc[i,'Lugar']}\n"+'|'+f"{data.loc[i,'Fecha']}\n"
HTML_results+='|}'
#print(HTML_results)
with open('resultados.txt', 'w') as file:
        file.write (HTML_results)
```
