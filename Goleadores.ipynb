```python
import pandas as pd
import numpy as np
import re
```

```python
data_location = r"C:\Users\Vander\Desktop\A Mano\Proyecto\test.csv"

data = pd.read_csv(data_location)
data.Goleadores1 = data.Goleadores1.replace(" y ", ", ")
print(data.Goleadores1)
```

```python
lista=[]
goles=0
df=pd.DataFrame({"Goleadores1": data.Goleadores1, "Goleadores2": data.Goleadores2})
for Equipo in df.index:
    g1=str(df.loc[Equipo,'Goleadores1']).replace(" y ", ", ")
    g1=g1.replace(" (p)", "")
    g1=g1.replace(" (1p)", "")
    g1=g1.replace(" (2p)", "")
    g1=g1.replace(" (3p)", "")
    g1=g1.replace("nan", "")
    g1=g1.split(", ")
    lista.append(g1)
    #print(g1)
df2=pd.DataFrame({"Nombre":lista,
                "Goles":goles})
df2
```

```python
pattern=r'((\D*)(\d*)?(, ))*'
for x in lista:
    x+=", "
    gld=re.findall(pattern, x)
    print(gld)
```
